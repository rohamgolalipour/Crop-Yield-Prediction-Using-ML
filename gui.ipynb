```python
import pandas as pd
import tkinter as tk
import joblib
```

```python
reg = joblib.load('regressor.joblib')
cls = joblib.load('classifier.joblib')
```

```python
root = tk.Tk()
root.geometry('450x550')
root.eval('tk::PlaceWindow . center')
```

```python
default = 'Choose'
```

```python
def predict():
    input_df = pd.DataFrame({
        'moisture': [float(moist_.get())],
        'temp': [float(temp_.get())],
        'sunlight': [int(light_.get())],
        'fertilizer': [int(fert_.get())],
        'rainfall': [int(rain_.get())],
        'size': [int(size_.get())],
        'crop': [crop_.get()],
        'irrigation': [irri_.get()],
    })

    reg_pred = reg.predict(input_df)[0]
    cls_pred = cls.predict(input_df)[0]

    if cls_pred == 1:
        suitness = 'suitable'
    else:
        suitness = 'unsuitable'

    result_label.config(text = f"According to our analysis,\nyour farm is expected to produce approximately {reg_pred:.1f} tons of {input_df['crop'][0]} per hectare.\nAnd your farm is {suitness} based on current conditions.")
```

```python
tk.Label(root, text = 'Soil Moisture:').pack(pady = 3)
moist_ = tk.Entry(root)
moist_.pack()

tk.Label(root, text = 'Average Temperature:').pack(pady = 3)
temp_ = tk.Entry(root)
temp_.pack()

tk.Label(root, text = 'Sunlight:').pack(pady = 3)
light_ = tk.Entry(root)
light_.pack()

tk.Label(root, text = 'Fertilizer:').pack(pady = 3)
fert_ = tk.Entry(root)
fert_.pack()

tk.Label(root, text = 'Rainfall:').pack(pady = 3)
rain_ = tk.Entry(root)
rain_.pack()

tk.Label(root, text = 'Size:').pack(pady = 3)
size_ = tk.Entry(root)
size_.pack()

tk.Label(root, text = 'Crop:').pack(pady = 3)
crop_ = tk.StringVar(value = default)
crop_menu = tk.OptionMenu(root, crop_, 'wheat', 'corn', 'cotton', 'rice')
crop_menu.pack()

tk.Label(root, text = 'Irrigation:').pack(pady = 3)
irri_ = tk.StringVar(value = default)
irri_menu = tk.OptionMenu(root, irri_, 'low', 'medium', 'high')
irri_menu.pack()

tk.Button(root, text = 'See Results:', command = predict).pack(pady = 10)

result_label = tk.Label(root, text = '')
result_label.pack(pady = 10)
```

```python
root.mainloop()
```
